```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_excel("Downloads\dados-12-14.xls")
```

```python
dados2 = dados.set_index("dateRep")
dados2 = dados2.sort_index()
sns.set(rc={'figure.figsize':(20,10)})
```

```python
dados2["mortes"] = dados2.groupby("countriesAndTerritories")["deaths"].cumsum()
dados3 = dados2.pivot(columns="countriesAndTerritories")
dados3["mortes"].plot()
```

```python
dados2["Casos_Hab"] = dados2.groupby("countriesAndTerritories")["cases"].cumsum()/1000
dados4 = dados2.pivot(columns="countriesAndTerritories")
dados4["Casos_Hab"].plot()
```
